```python
plt.figure(figsize=(10,10), dpi=400)
# plt.scatter(df['Product_Price'],df['quantity_sold'],color='#7fff50')
# plt.plot(kind='bar', ax=df['Product_Fat_Content'])
plt.hist(('Cluster 1', 'Cluster 2', 'Cluster 3'), weights = locAndType.loc[(slice(None), slice('Supermarket Type1','Supermarket Type1')), :]['Product_Supermarket_Sales'],color='#7fff50')
plt.xlabel('Supermarket Type 1')
plt.ylabel('Product Supermarket Sales')
plt.title('Comparison of sales for Supermarket Type 1 in different clusters')
plt.savefig('Sales for supermarket type1 in all clusters.png')
```
