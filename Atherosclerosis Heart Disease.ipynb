```python
df = df.dropna()
```

```python
pd.unique(df['Thal'])
```

```python
df['Thal'] = df['Thal'].replace('fixed',1)
df['Thal'] = df['Thal'].replace('normal',2)
df['Thal'] = df['Thal'].replace('reversable',3)
```

```python
df.head()
```

```python
df.info()
```

```python
df['AHD'] = df['AHD'].replace('Yes',1)
df['AHD'] = df['AHD'].replace('No',0)
```

```python
df.head()
```

```python
df = pd.concat([df, pd.get_dummies(df['ChestPain'], 'CPtype')], axis=1)
df = pd.concat([df, pd.get_dummies(df['RestECG'], 'ECG')], axis=1)
df = pd.concat([df, pd.get_dummies(df['Thal'], 'thal')], axis=1)
df = pd.concat([df, pd.get_dummies(df['Slope'], 'peakSlope')], axis=1)
```

```python
df = df.drop(['Unnamed: 0', 'ChestPain', 'RestECG', 'Thal', 'Slope'], axis=1)
df
```

```python
plt.figure(figsize = (18,8), dpi=300)
sns.heatmap(df.corr(), annot=True,)
```

```python
X = df[['Age', 'Sex', 'RestBP', 'Chol', 'MaxHR', 'ExAng', 'Oldpeak',
       'Ca', 'CPtype_asymptomatic', 'ECG_1','thal_3', 'peakSlope_2']]
y = df['AHD']
```

```python
df.columns
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=101,test_size=0.3)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,predictions) * 100
```
